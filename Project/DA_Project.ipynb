**Load the Data**
First, load the data into a Pandas DataFrame.

```python
import pandas as pd
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("book_sales.db")


# Load data into a DataFrame
query = "SELECT * FROM book_sales"

# Turn SQL Query to DataFrame (<query>, <source of data / database>)
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()


```

```python
# recall dataframe df
df 
```

## **Data Manipulation**

```python
# Display the first three rows of the DataFrame
df.head(3)
```

```python
# Check for missing value
df.isnull().sum()

```

```python
# Calculate total sales by multiplying quantity sold by price per copy
df["total sales"] = df["quantity_sold"] * df["price_per_copy"]
df
```

```python
# Total revenue by genre
df.groupby("genre")["total sales"].sum()




```

```python
# Average revenue per genre
df.groupby("genre")["total sales"].mean()

```

```python
# Standard deviation of revenue per genre
df.groupby("genre")["total sales"].std()
```

```python
# Overall mean and standard deviation of total sales
df["total sales"].describe()[["mean", "std"]]
```

```python
# Group by author and sum total sales
df.groupby("author")["total sales"].sum(())
```

```python
# Mean quantity sold per genre
df.groupby("genre")["quantity_sold"].mean()
```

```python
# Standard deviation of quantity sold per genre
df.groupby("genre")["quantity_sold"].std()
```

```python
# Display unique authors
unique_authors = pd.DataFrame(df["author"].unique(), columns= ["Authors"])
unique_authors



```

```python
# Top 5 authors by total sales
# Note: use .nlargest(5) method after sum() to get the top 5
df.groupby("author")["total sales"].sum().sort_values(ascending= False).head(5)

```

```python
# Describe or get the distribution of total sales across all genres
df.groupby("genre")["total sales"].describe()
```

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt



# Print the total revenue by genre
df.groupby("genre")["total sales"].sum()


# Generating a pie chart
plt.pie(df.groupby("genre")["total sales"].sum(), colors = ["red", "blue", "pink"], autopct= "%.2f%%", labels= df["genre"].unique())


# Set the title of the plot
plt.title("Total Revenues Per Genre")

# Display the plot
plt.show()
```

```python
plt.bar (df["book_title"], df["total sales"])

plt.xlabel ("Book Titles")
plt.ylabel ("Total Sales")
plt.xticks(rotation = 90, ha = "center", fontsize = 8)

plt.show()
```

```python
df.groupby("book_title")["total sales"].sum().sort_values(ascending=False).head(5)
```

```python
plt.bar(df["book_title"] , df["quantity_sold"])
plt.xlabel("Book Titles")
plt.ylabel("Quantity Sold")
plt.xticks(rotation = 90, ha = "center")
plt.show()
```

```python
df.groupby("book_title")["quantity_sold"].sum().sort_values(ascending=False).head(5)
```
