## **Introduction to Data Visualization**

`Data visualization` is the process of presenting information and data in a graphical format. Utilizing visual elements such as charts, graphs, and maps, data visualization tools offer an intuitive approach to interpreting trends, outliers, and patterns within data.

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd 

# Created a dataframe with sales data

df = pd.DataFrame({
    "Month" : ["Jan" , "Feb", "March", "Apr", "May"],
    "Sales" : [1200,1500,1700,1800,1600]
})

# Plot the data points for the (Month, Sales)
plt.plot(df["Month"], df["Sales"], marker = "x")

#Set the title of the plot
plt.title("Monthly Sales Data")

# Label the x-axis
plt.xlabel("Month")

# Label the y-axis
plt.ylabel("Sales ($)")

# Display the plot
plt.show()
```

### Types of Charts
- Bar
- Pie
- Line

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd

# Generate a bar chart
plt.bar(df["Month"], df["Sales"] , color = "red")

# Set the title of the plot
plt.title ("Monthly Sales Data")

# Label the x - axis
plt.xlabel("Month")

# Label the y - axis
plt.ylabel("Sales ($)")

# Display the bar chart
plt.show()
```

```python
# Import the matplotlib.pyplot module
import matplotlib.pyplot as plt
import pandas as pd

# Generating the Pie Chart
plt.pie (df["Sales"], labels = df["Month"], colors = ["tomato", "lightgreen", "skyblue", "red", "purple"], autopct= "%.2f%%")

# Set the title of the plot
plt.title ("Monthly Sales Data")

# Display the plot
plt.show()
```
